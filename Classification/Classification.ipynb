# Data Preprocessing

```python
# Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# importing database
dataset = pd.read_csv('../Dataset/breast_cancer.csv')
```

```python
#checking the head
dataset.head()
```

```python
dataset.info()
```

```python
import seaborn as sns
sns.pairplot(dataset)
```

```python
# Splitting the dependent and independent variables
x = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values.reshape(-1, 1)
```

```python
# checking the values
print(x)
```

```python
print(y)
```

```python
# splitting the dataset into training and test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y.reshape(-1, 1), test_size=0.2)
```

# Training the model

```python
# using logistic regression
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
print(np.concatenate((y_test.reshape(-1, 1), y_pred.reshape(-1,1)), 1))
```

# Evaluating the model

```python
# confusion matrixcs
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
# cross-validation
from sklearn.model_selection import cross_val_score
```

```python
cvs = cross_val_score(estimator=lr, X=x_train, y=y_train.reshape(len(y_train), 1), cv = 10)
```

```python
cvs.mean()*100
```

```python
np.std(cvs)*100
```
